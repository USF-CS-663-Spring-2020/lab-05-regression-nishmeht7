```
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression

df = pd.read_csv('Credit.csv')
df
```

INCOME against LIMIT

```
df.plot(kind='scatter', x='Income', y='Limit')
```

This looks homogenous

```
plot = sns.residplot(x='Income', y='Limit', data=df)
plot.set(xlabel='Income', ylabel='Limit')
```

```
lin_reg = LinearRegression()
df_1 = df.iloc[:, 1:3]
lin_reg.fit(df_1.drop('Limit', axis=1), df_1.iloc[:,1])
lin_reg
```

```
lin_reg.coef_
```

```
lin_reg.intercept_
```

SLR can be applied giving us a coefficient of 51.875 and an intercept of 2389.8694

**RATING against LIMIT**

```
df.plot(kind='scatter', x='Rating', y='Limit')
```

```
plot1 = sns.residplot(x='Rating', y='Limit', data=df)
```

```
lin_reg1 = LinearRegression()
df_2 = df.iloc[:, [2,3]]
lin_reg1.fit(df_2.drop('Limit', axis=1), df.iloc[:,0])
lin_reg1
```

```
lin_reg1.coef_
```

```
lin_reg1.intercept_
```

Since the data is homoegenous we can apply SLR with a coefficient of 0.016 and an intercept of 194.6689
